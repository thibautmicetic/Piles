```python
class Pile:
    def __init__(self):
        self._mapile = []
    
    def __len__(self):
        return len(self._mapile)
    
    def sommet(self):
        if len(self._mapile) > 0:
            return self._mapile[0]
    
    def empiler(self, element):
        self._mapile += [element]
    
    def depiler(self):
        if len(self._mapile) > 0:
            return self._mapile.pop(0)
```

```python
p = Pile()
p.empiler(3)
p.empiler(5)
print(p.sommet())
p.empiler(8)
p.depiler()
print(p.depiler())
print(p.depiler())
```
