```python
class Pile:
    def __init__(self):
        self.mapile = []
    
    def __len__(self):
        return (len(self.mapile))
    
    def sommet(self):
        if len(self.mapile) > 0:
            return self.mapile[0]
    
    def empiler(self, element):
        self.mapile = [element] + self.mapile
    
    def depiler(self):
        if len(self.mapile) > 0:
            return self.mapile.pop(0)
```

```python
p = Pile()
p.empiler(3)
p.empiler(5)
print(p.sommet())
p.empiler(8)
p.depiler()
print(p.depiler())
print(p.depiler())
```
